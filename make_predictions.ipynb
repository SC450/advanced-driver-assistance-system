# Real Time Object Detection!

```python
import matplotlib.pyplot as plt

img = cv.imread("test_image_2.jpg")
pred_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)  # Convert BGR to RGB

plt.imshow(pred_img)
plt.axis("off")
plt.show()
```

```python
# Get the video
vid = cv.VideoCapture("dashcam_data/16.mp4")

# Start the video
while True:
    ret, frame = vid.read()

    # Check if we cannot read the video
    if not ret:
        print("Unable to read video.")
        break

    # Else, read the video
    pred_image = yolo.predictions(frame)
    cv.namedWindow("YOLO Predictions", cv.WINDOW_NORMAL)
    cv.imshow("YOLO Predictions", pred_image)

    # End the video
    if cv.waitKey(1) == 27:
        break

cv.destroyAllWindows()
vid.release()
```
